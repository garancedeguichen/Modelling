# What is Logistic Regression?
Logistic Regression is a statistical method used for analyzing a dataset where the dependent variable (also known as the target variable) is categorical. Specifically, logistic regression is well-suited for binary classification problems, where the target variable can take on one of two possible outcomes, often represented as 0 or 1, True or False, or positive or negative.

Conceptually, logistic regression is an extension of linear regression. While linear regression is used to model the relationship between one or more independent variables and a continuous dependent variable, logistic regression models the relationship between one or more independent variables and the probability of an event occurring (i.e., the dependent variable taking on a particular value).



# Applications of Logistic Regression:

- Medical Diagnostics: Logistic regression can be used to predict the likelihood of a patient having a certain disease or condition based on a set of diagnostic test results or risk factors.
- Finance: In the banking industry, logistic regression can be used to predict the probability of a customer defaulting on a loan, which helps banks make lending decisions.
- Marketing: Logistic regression can be employed to predict customer churn or conversion rates, which allows companies to focus their marketing efforts more effectively.
- Human Resources: Logistic regression can be applied to predict the likelihood of an employee leaving the company, enabling proactive retention efforts.
- Fraud Detection: In security applications, logistic regression can help predict fraudulent activities, such as credit card fraud or insurance fraud, by identifying suspicious patterns in the data.
- Political Science: Logistic regression can be used to analyze and predict voter preferences based on demographic and social variables.

In summary, logistic regression is a powerful and widely-used statistical method that allows us to model the relationship between one or more independent variables and a binary outcome. It is especially useful in classification problems and has a variety of real-world applications across different industries.

# Logistic vs Linear
Logistic Regression and Linear Regression are two widely used regression techniques in machine learning and statistics. Although they share some similarities, they are primarily used for different types of problems and have different underlying mechanisms. Here is a comparison between the two:

###Problem Type:

- Linear Regression: It is used when the target variable (dependent variable) is continuous. Examples include predicting house prices, stock prices, or a student's test score.
- Logistic Regression: It is used when the target variable is categorical, particularly for binary classification problems, where the target variable can take one of two possible outcomes. Examples include predicting whether a customer will churn, whether an email is spam, or whether a transaction is fraudulent.

###Relationship between Variables:

- Linear Regression: Assumes a linear relationship between the independent variables and the target variable. The goal is to find the best-fitting straight line that minimizes the sum of squared residuals between the actual and predicted values.
- Logistic Regression: Models the probability of an event occurring, i.e., the target variable taking a particular value. It employs the logistic (sigmoid) function to map the relationship between the independent variables and the target variable, producing an S-shaped curve.

###Output:

- Linear Regression: Predicts a continuous value that can range from negative infinity to positive infinity.
- Logistic Regression: Predicts the probability of an event, which is bounded between 0 and 1. To make a binary classification decision, a threshold is applied to the predicted probabilities (e.g., 0.5), resulting in a prediction of either 0 or 1, True or False, or positive or negative.

###Loss Function:

- Linear Regression: The most common loss function used is the Mean Squared Error (MSE), which calculates the average of the squared differences between the actual and predicted values.
- Logistic Regression: Utilizes the log loss (cross-entropy) as the loss function, which measures the difference between the actual probabilities (0 or 1) and the predicted probabilities.

###Evaluation Metrics:

- Linear Regression: Metrics like Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE) are commonly used to evaluate the performance of the model.
- Logistic Regression: Classification metrics such as accuracy, precision, recall, F1 score, and Area Under the Receiver Operating Characteristic (ROC) Curve (AUC-ROC) are employed to assess the model's performance.


In summary, linear regression is designed for continuous target variables, while logistic regression is tailored for binary classification problems. They have different assumptions about the relationship between variables, produce different outputs, and employ distinct loss functions and evaluation metrics.

# Logistic (Sigmoid) Function

The logistic function, also known as the sigmoid function, is an essential component of logistic regression. It is a mathematical function that maps any real-valued input to a value between 0 and 1, producing an S-shaped curve. This characteristic makes the logistic function well-suited for modeling probabilities, as probabilities must be bounded between 0 and 1.

The logistic function is defined as follows:

$$
f(x) = \frac{1}{1 + e^{-x}}
$$

where x is the input, e is the base of the natural logarithm (approximately 2.718), and f(x) is the output of the logistic function.

The key properties of the logistic function are:

- Sigmoid shape: The logistic function produces an S-shaped curve, which starts at 0, increases gradually, and then asymptotically approaches 1.
- Range: The output of the logistic function is bounded between 0 and 1 (0 < f(x) < 1), making it suitable for modeling probabilities.
- Monotonicity: The logistic function is strictly monotonically increasing, which means that as the input value increases, the output value also increases.
- Centered at 0: The logistic function is symmetric around the input value 0, where f(0) = 0.5. Input values greater than 0 produce outputs larger than 0.5, while input values less than 0 yield outputs smaller than 0.5.
<br>

In the context of logistic regression, the logistic function is used to map the linear combination of input features (independent variables) to the probability of an event (dependent variable) occurring. This probability is then used to make a binary classification decision by applying a threshold (typically 0.5).

Here's an example of how the logistic function is used in logistic regression:

Let's say we have a linear combination of input features as follows:

$$
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
$$


where Î²i are the coefficients (weights) and xi are the input features.

The logistic function is then applied to z:

$$
p = \frac{1}{1 + e^{-z}}
$$


where p is the output probability, which represents the probability of the target variable taking a specific value (e.g., 1). By applying a threshold to p, we can make a binary classification decision.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generating synthetic data
# np.random.seed(0)
x1 = np.random.normal(50, 10, 100)  # Exam 1 scores
x2 = np.random.normal(50, 10, 100)  # Exam 2 scores
X = np.vstack((x1, x2)).T

# Admission decisions (randomized for illustration)
y = (x1 + x2 > 100).astype(int)  # Admitted if the sum of scores is more than 100

# Scatter plot of the synthetic data
plt.figure(figsize=(10, 6))
plt.scatter(x1[y == 1], x2[y == 1], color='b', label='Admitted')
plt.scatter(x1[y == 0], x2[y == 0], color='r', label='Not Admitted')
plt.xlabel('Exam 1 Score')
plt.ylabel('Exam 2 Score')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generating synthetic data
# np.random.seed(0)
x1 = np.random.normal(50, 10, 100)  # Exam 1 scores
x2 = np.random.normal(50, 10, 100)  # Exam 2 scores
X = np.vstack((x1, x2)).T

# Admission decisions (randomized for illustration)
noise = np.random.normal(0, 5, 100)
y = (x1 + x2 + noise > 100).astype(int)
# y = (x1 + x2 + np.random.randint(-10, 10) > 100).astype(int)  # Admitted if the sum of scores is more than 100

# Creating and training a logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Decision boundary calculation
x_values = np.array([np.min(x1), np.max(x1)])
y_values = -(model.intercept_ + np.dot(model.coef_[0][0], x_values)) / model.coef_[0][1]

# Scatter plot of the synthetic data
plt.figure(figsize=(10, 6))
plt.scatter(x1[y == 1], x2[y == 1], color='b', label='Admitted')
plt.scatter(x1[y == 0], x2[y == 0], color='r', label='Not Admitted')

# Adding the decision boundary to the plot
plt.plot(x_values, y_values, label='Decision Boundary', color='green')

plt.xlabel('Exam 1 Score')
plt.ylabel('Exam 2 Score')
plt.title('Exam Scores and Admission Decision')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for _ in range(10):  # Plotting 10 different decision boundaries
    noise = np.random.normal(0, 5, 100)
    y_temp = (x1 + x2 + noise > 100).astype(int)
    model_temp = LogisticRegression().fit(X, y_temp)
    y_values_temp = -(model_temp.intercept_ + np.dot(model_temp.coef_[0][0], x_values)) / model_temp.coef_[0][1]
    plt.plot(x_values, y_values_temp, color='grey', alpha=0.5)  # Plot in grey with some transparency

# Adding the original decision boundary for comparison
plt.plot(x_values, y_values, label='Decision Boundary', color='green')
plt.scatter(x1[y == 1], x2[y == 1], color='b', label='Admitted')
plt.scatter(x1[y == 0], x2[y == 0], color='r', label='Not Admitted')
plt.xlabel('Exam 1 Score')
plt.ylabel('Exam 2 Score')
plt.title('Impact of Random Noise on Admission Decision')
plt.legend()
plt.show()
```

###Confusion Matrix:
A confusion matrix is a table that displays the true positive (TP), true negative (TN), false positive (FP), and false negative (FN) predictions of a classifier. It helps to visualize the classifier's performance and identify where it is making mistakes.

###Accuracy:
Accuracy is the ratio of the correct predictions (TP and TN) to the total number of predictions (TP, TN, FP, and FN). It measures the overall performance of the classifier.
Formula: Accuracy = (TP + TN) / (TP + TN + FP + FN)

###Precision:
Precision, also known as positive predictive value, is the ratio of true positive predictions to the total number of positive predictions (TP and FP). It measures the classifier's ability to correctly identify positive instances.
Formula: Precision = TP / (TP + FP)

###Recall:
Recall, also known as sensitivity or true positive rate, is the ratio of true positive predictions to the total number of actual positive instances (TP and FN). It measures the classifier's ability to find all the positive instances.
Formula: Recall = TP / (TP + FN)

###F1 Score:
The F1 score is the harmonic mean of precision and recall. It is a single metric that combines both precision and recall, providing a balanced measure of a classifier's performance. The F1 score is particularly useful when dealing with imbalanced datasets or when both false positives and false negatives are important to consider.
Formula: F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

## Introduction to Time Series

A time series is a sequence of data points collected or recorded at regular or irregular intervals over a period. Time series data is unique because the order of data points is crucial, as each observation depends on the time it was collected. Analyzing time series data can help uncover patterns, trends, and seasonality within the data, as well as make forecasts for future values.

## Importance and Applications of Time Series Analysis:

Time series analysis has a wide range of applications across various domains, including:

- Finance and economics: Stock prices, exchange rates, GDP, and inflation rates
- Environmental studies: Temperature, rainfall, and air quality measurements
- Healthcare: Patient vital signs, disease prevalence, and hospital admissions
- Retail: Sales data, inventory levels, and consumer demand
- Industrial: Production data, equipment maintenance, and quality control

In these applications, time series analysis helps organizations make data-driven decisions, optimize processes, and anticipate future trends.

## Components of a Time Series

A time series can typically be broken down into three main components: trend, seasonality, and noise (also referred to as residuals or irregular fluctuations). Understanding these components is crucial for selecting the appropriate models and methods for time series analysis.

**Trend**: The long-term pattern or direction of the data. A trend can be upward, downward, or constant over time. Identifying the trend can help us understand the general behavior of the time series and forecast future values.

Trend is calculated using a centered moving average of the time series. The moving average is calculated using a window length corresponding to the frequency of the time series. For example, we would use a window of length 12 for monthly data.

Smoothing the series using such a moving average comes together with some disadvantages. First, we are âlosingâ the first and last few observations of the series. Second, the MA tends to over-smooth the series, which makes it less reactive to sudden changes in the trend or jumps.

**Seasonality**: The periodic fluctuations that occur in the data due to seasonal factors. Seasonality can be observed in daily, weekly, monthly, or yearly patterns. For instance, sales of winter clothes tend to peak during the winter months, while air conditioner sales might peak during the summer.

To calculate the seasonal component, we first need to detrend the time series. We do it by subtracting the trend component from the original time series (remember, we divide for the multiplicative variant).

Having done that, we calculate the average values of the detrended series for each seasonal period. In the case of months, we would calculate the average detrended value for each month.

The seasonal component is simply built from the seasonal averages repeated for the length of the entire series Again, this is one of the arguments against using the simple seasonal decomposition â the seasonal component is not allowed to change over time, which can be a very strict and often unrealistic assumption for longer time series.

**Noise**: The irregular or random fluctuations in the data that cannot be explained by the trend or seasonality. Noise is often the result of short-term, unpredictable events or measurement errors.

It is simply what is left after removing (by subtracting or dividing) the trend and seasonal components from the original time series.

## Stationarity in Time Series

Stationarity is a key concept in time series analysis because many statistical models, including time series forecasting models, rely on the assumption that the underlying data is stationary. A stationary time series is one whose statistical properties, such as mean, variance, and autocorrelation, do not change over time.

Importance of stationarity in time series analysis:
Stationary data allows the model to make accurate predictions because the underlying properties of the data remain constant over time.
Non-stationary data can lead to inaccurate or unreliable predictions because the statistical properties of the data change over time.
Many forecasting models, including ARIMA and its variations, assume that the input time series is stationary.

Types of stationarity:

1. Strong stationarity: A time series is strongly stationary if its joint probability distribution is invariant under time shifts. In other words, the statistical properties of the series remain the same regardless of the time period.
2. Weak stationarity (also called covariance stationarity): A time series is weakly stationary if its mean, variance, and autocovariance (the covariance between data points at different time lags) are constant over time.
Weak stationarity is more commonly used in time series analysis because it is easier to test and more practical for real-world data.

Common properties of stationary time series:
- Constant mean: The average value of the time series remains the same over time.
- Constant variance: The variability of the time series does not change over time.
- Constant autocorrelation: The relationship between data points at different time lags remains consistent over time.

To transform a non-stationary time series into a stationary one, various techniques such as differencing, log transformations, or seasonal adjustments can be applied. It's important to note that some time series, such as those with structural breaks or regime changes, may never be fully stationary. In such cases, alternative models that do not require stationarity assumptions may be more appropriate.

In summary, understanding stationarity in time series is essential because many statistical models rely on the assumption of stationarity. A stationary time series is one whose statistical properties remain constant over time, which allows the model to make accurate and reliable predictions. Transforming non-stationary data into stationary data is often a crucial step in time series analysis and forecasting.

## Testing for Stationarity: The Augmented Dickey-Fuller:



```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    result = adfuller(series)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'{key}: {value}')

# Perform the ADF test on the time series data
adf_test(data_filled['TEMP_C'])
```

The ADF statistic of -4.1658 is more negative than the critical values at all three significance levels (1%, 5%, and 10%). This means we can reject the null hypothesis of non-stationarity with a high degree of confidence. The p-value of 0.0007525 also supports this conclusion, as it is much smaller than the typical threshold of 0.05 for rejecting the null hypothesis. Based on these results, we can conclude that the given time series is likely stationary.
